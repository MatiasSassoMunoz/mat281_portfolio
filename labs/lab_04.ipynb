EL conjunto de datos se denomina `Automobile_data.csv`, el cual contiene información tal como: compañia, precio, kilometraje, etc.

Lo primero es cargar el conjunto de datos y ver las primeras filas que lo componen:

```python
import pandas as pd
import numpy as np
import os
```

```python
# cargar datos
df = pd.read_csv(os.path.join("data","Automobile_data.csv")).set_index('index')
df.head()
```

El objetivo es tratar de obtener la mayor información posible de este conjunto de datos. Para cumplir este objetivo debe resolver las siguientes problemáticas:

1. Elimine los valores nulos (Nan)

```python
mask = lambda df: df.notnull().all(axis=1)
df = df[mask]
```

2. Encuentra el nombre de la compañía de automóviles más cara

```python
#se considerará la compañía de automóviles más cara 
#a la que tenga el mayor promedio de precios
 
marca = df.groupby('company')  
marca_precio = marca.agg({'price':[np.mean]}).reset_index()
print(marca_precio)
preciomax = float(marca_precio['price'].max())

n = len(marca_precio['company'])
    
diccio = {}
    
for k in range(n):
    
    diccio[k] = [0,0]              #se crea un diccionario para guardar un índice como llave y         
                                   #como valor de cada llave, una lista con el nombre de la compañia en 
                                   #la primer entrada y en la segunda entrada
for i in range(n):                 #se guarda el valor de promedio de precios correspondiente a la compañía  
    
    diccio[i][0] = marca_precio['company'][i]    #se guarda cada compañía en la primera entrada de la lista de cada compañía 
    
cont=0

for k in marca_precio['price']['mean']:
        diccio[cont][1]=k                 #se guarda el promedio de precios de la compañía correspondiente 
        cont+=1
        
for i in diccio.values():                #finalmente se encuentra a la compañía que tiene el mayor valor de promedio de precios
    if i[1]==preciomax:                      
        print('\n La compañía de automóviles más cara es:',i[0])
        
```

3. Imprimir todos los detalles de Toyota Cars

```python
diccio3 = {}
diccio3['toyota'] = []
diccio4 = dict(df['company'])                
lista4=list(diccio4.keys())

for k in lista4:
    if diccio4[k]=='toyota':
        diccio3['toyota'].append(k)       #se buscan las posiciones de los autos toyota y se guardan en la lista 
                                          #con llave 'toyota'
lista5=[]                                 
for k in diccio3['toyota']:              #se agregan las filas de autos toyota en la lista 'lista5' (como dataframes)   
    lista5.append(df.loc[k])

lista6=[]
for frame in lista5:
    lista=list(frame)                     #se convierten los dataframes en listas 
    lista6.append(lista)
    
arreglo=np.array(lista6)                   #la lista obtenida se transforma a un arreglo de numpy
    
dataframe=pd.DataFrame(arreglo)            #se ocupa este arreglo para generar un nuevo dataframe con las filas del original
dataframe.columns=['company','body-style','wheel-base','lenght','engine-type','num-of-cylinders','horsepower','average-mileage','price','cantidad de autos']
#aquí se cambian los títulos de las columnas por los correctos

dataframe.describe(include='all') #finalmente se hace una descripción del dataframe que solo contiene informacion de 
                                  #de la compañia Toyota 

```

4. Cuente el total de automóviles por compañía

```python
companias=df
companias['cantidad de autos']=1         #se agrega una columna con 1's en sus entradas 
companias2=companias.groupby('company')
companias3=companias2.agg({'cantidad de autos':[sum]}).reset_index()
companias3
```

5. Encuentra el coche con el precio más alto por compañía

```python
marca=df.groupby('company')
marca_max=marca.agg({'price':[np.max]})
marca_max
```

6. Encuentre el kilometraje promedio (**average-mileage**) de cada compañía automotriz

```python
compañia=df.groupby('company')
compañia_kilometrajeprom = compañia.agg({'average-mileage':[np.mean]})
compañia_kilometrajeprom
```

7. Ordenar todos los autos por columna de precio (**price**)

```python
dF3=df.sort_values('price').reset_index()
dF3
```
